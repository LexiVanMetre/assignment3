5. Treat this as a classification problem: The model is supposed to predict 0 or 1 for each customer, classifying them as a purchaser or non-purchaser. Use sklearn to determine the optimal number of neighbors $k$ to use.


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Determining the optimal number k:
k_bar = 30
k_grid = np.arange(1, k_bar)
SSE = np.zeros(k_bar)
for k in range(k_bar):
  model = KNeighborsClassifier(n_neighbors=k+1)
  fitted_model = model.fit(X_train, y_train)
  y_hat = fitted_model.predict(X_test)
  SSE[k] = np.sum( (y_test - y_hat)**2 )

SSE_min = np.min(SSE)
min_index = np.where(SSE==SSE_min)
k_star = k_grid[min_index]
print(k_star)
````

````python
plt.plot(np.arange(0, k_bar), SSE, label='Test')
plt.xlabel("k")
plt.ylabel("SSE")
plt.legend(loc='lower right')
plt.title('SSE')
plt.show()
````

> It looks like the optimal k is at 6, 16 according to the k_star.

4. Use the $k$NN algorithm for regression and the training data to predict `price` using `year` and `mileage` for the test set for $k=3,10,25,50,100,300$. For each value of $k$, compute the Sum of Squared Error and make a scatterplot showing the test value plotted against the predicted value. What patterns do you notice as you increase $k$?


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

k_values = [3, 10, 25, 50, 100, 300]
SSE = []

for k in k_values:
    model = KNeighborsRegressor(n_neighbors=k)
    fitted_model = model.fit(X_train, y_train)
    y_hat = fitted_model.predict(X_test)
    SSE.append(mean_squared_error(y_test, y_hat))

plt.figure(figsize=(8, 6))
for i, k in enumerate(k_values):
    plt.subplot(2, 3, i + 1)
    plt.scatter(y_test, fitted_model.predict(X_test), label=f'k={k}')
    plt.title(f'k={k}')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.legend()
plt.tight_layout()
plt.show()
````

````python
plt.plot(k_values, SSE, marker='o')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('Sum of Squared Error (SSE)')
plt.title('SSE vs. k for k-NN Regression')
plt.show()
````

I do not see any difference between the scatterplots for the predicted price versus actual price in the graphs between k=3 to k = 300. I think this is because the optimal k value is 2.

---
However, I do see as we increase the k value the sum of squared error decreases. As you increase k, the predictions tend to become smoother. This is because with a larger k, the algorithm considers more neighboring data points when making predictions. The predicted values will be more of an average or consensus of the neighboring points. Therefore it makes sense that the sum of the squared errors becomes less as you have more predictors.

**Q6.** This is a question about $k$ means clustering. We want to investigate how adjusting the "noisiness" of the data impacts the quality of the algorithm and the difficulty of picking $k$.

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.

````python
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?


````python
import seaborn as sns
from sklearn.cluster import KMeans # Import kmc
import matplotlib.pyplot as plt
sns.scatterplot(data = df0_125, x = 'x1', y = 'x2', hue = 'group', style = 'group')
````

````python
sns.scatterplot(data = df0_25, x = 'x1',y='x2',hue='group',style='group')
````

````python
sns.scatterplot(data = df0_5, x = 'x1',y='x2',hue='group',style='group')
````

````python
sns.scatterplot(data = df1_0, x = 'x1',y='x2',hue='group',style='group')
````

````python
sns.scatterplot(data = df2_0, x = 'x1',y='x2',hue='group',style='group')
````

> As the noise increases, the less uniformed the groups of clustered points are. At noise level 1.0, the distinctness of the clusters is no longer visible.

3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$.


````python
def maxmin(x):
  x = (x-min(x))/(max(x)-min(x))
  return x
def scree(data):
  X = data.loc[ : , ['x1','x2'] ]
  X = X.apply(maxmin)

# Create a scree plot:
  k_bar = 15
  k_grid = np.arange(1,k_bar+1)
  SSE = np.zeros(k_bar)
  for k in range(k_bar):
    model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0)
    model = model.fit(X)
    SSE[k] = model.inertia_
  scree_plot, axes = plt.subplots()
  sns.lineplot(x=k_grid, y=SSE).set_title('Scree Plot')
  axes.set_ylim(0, 35)
````

````python
scree( data = df0_125 )
````

> There is an elbow at k=3.

````python
scree(data = df0_25)
````

> There is an elbow at k=3.

````python
scree(data = df0_5)
````

> There is an elbow at k=3.

````python
scree(df1_0)
````

> Maybe an elbow at k=2.

````python
scree(df2_0)
````

> The curve is mostly smooth, not sure where the position of the elbow is.

4. Explain the intuition of the elbow, using this numerical simulation as an example.

> When the groups are very different, the marginal benefit of adding another cluster is larger since it will decrease the SSE. As variation and noise increases, the groups spread out, and eventually they are distinct. Even if an observer would be able to tell a point belongs to a certain cluster, the more the points overlap the less likely one can observe specific groups and the number of specific groups to create. When the scree plot has a distinct elbow, there are also visually separate clusters. However, when the clusters are just random points, the scree plot will be smooth.
